```python
#Combining the code
import pandas as pd
df=pd.read_csv('titanic.csv')

print("Missing values before cleaning:")
print(df.isnull().sum())

df_cleaned = df.dropna()
print(df_cleaned.isnull().sum())

df_cleaned.to_csv('titanic_cleaned.csv', index=False)
print("Cleaned dataset has been exported to 'titanic_cleaned.csv'.")
df_cleaned.head()


```

```python

import pandas as pd
import random

# Load the cleaned Titanic dataset
df_cleaned = pd.read_csv('titanic_cleaned.csv')

us_cities = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix", "Philadelphia", "San Antonio", "San Diego", "Dallas", "San Jose"]

random_cities = random.choices(us_cities, k=len(df_cleaned))

df_cleaned['city'] = random_cities

df_cleaned.to_csv('titanic_with_city.csv', index=False)

print("Cleaned dataset with city column has been exported to 'titanic_with_city.csv'.")
df_cleaned.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the cleaned Titanic dataset
df = pd.read_csv('titanic_cleaned.csv')

# Group data by sex and survival status and count the occurrences
survived_by_sex = df.groupby(['Sex', 'Survived'])['PassengerId'].count().reset_index()

# Create the bar plot using Seaborn
sns.barplot(x='Sex', y='PassengerId', hue='Survived', data=survived_by_sex)
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, ['No', 'Yes'], title='Survived')

plt.show()
```
